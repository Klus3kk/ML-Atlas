```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import FastICA
from sklearn.datasets import make_blobs
```

```python
# Generate synthetic data
X, _ = make_blobs(n_samples=500, centers=3, n_features=2, random_state=42)
```

```python
# Mix the data
A = np.array([[1, 1], [0.5, 2], [1.5, 0.5]])
X_mixed = np.dot(X, A.T)
```

```python
# Apply ICA
ica = FastICA(n_components=2)
X_ica = ica.fit_transform(X_mixed)
```

```python
# Plot the original and ICA-separated signals
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
plt.scatter(X_mixed[:, 0], X_mixed[:, 1], color='blue')
plt.title('Mixed Signals')
plt.xlabel('Mixed Feature 1')
plt.ylabel('Mixed Feature 2')

plt.subplot(1, 2, 2)
plt.scatter(X_ica[:, 0], X_ica[:, 1], color='red')
plt.title('ICA-Separated Signals')
plt.xlabel('ICA Feature 1')
plt.ylabel('ICA Feature 2')

plt.tight_layout()
plt.show()
```
